```python
sania_list = sania_df.to_dict(orient='records')

def CollectData(lists):
    WebDevelopmentList = []
    ProblemSolvingList = []
    EnjoyList = []
    OtherActivityList = []

    for item in lists:
        if 'localhost' in item['url'] or 'React' in item['url'] or 'nodejs' in item['url']:
            WebDevelopmentList.append(item)
        elif 'leetcode' in item['url'] or 'greekforce' in item['url']:
            ProblemSolvingList.append(item)
        elif 'youtube' in item['url']:
            EnjoyList.append(item)
        else:
            OtherActivityList.append(item)
    
    return WebDevelopmentList, ProblemSolvingList, EnjoyList, OtherActivityList

# Call the function with your sania_list
web_dev_list, problem_solving_list, enjoy_list, other_activity_list = CollectData(sania_list)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sania_df['date'] = pd.to_datetime(sania_df['date'])
sania_df['weekday'] = sania_df['date'].dt.day_name()
weekday_counts = sania_df.groupby('weekday')['url'].count().reset_index(name='Google_search')
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts['weekday'] = pd.Categorical(weekday_counts['weekday'], categories=weekday_order, ordered=True)
weekday_counts = weekday_counts.sort_values('weekday')
heatmap_data = pd.pivot_table(weekday_counts, values='Google_search', index='weekday', aggfunc='sum')
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt='g', cmap='YlGnBu', linewidths=.5)
plt.xlabel('Weekday')
plt.ylabel('Number of Google Searches')
plt.title('Google Searches by Weekday')
plt.show()
```
